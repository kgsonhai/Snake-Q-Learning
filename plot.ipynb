```python
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file output.txt
file_path = "output.txt"
rolling_averages = []

with open(file_path, "r") as file:
    for line in file:
        try:
            rolling_avg = float(line.strip())
            rolling_averages.append(rolling_avg)
        except ValueError:
            continue

# Vẽ biểu đồ
plt.plot(rolling_averages, label="Rolling Average Score")
plt.title("Rolling Average Score")
plt.xlabel("Game")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```
